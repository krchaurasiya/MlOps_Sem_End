```python
!pip install pandas numpy scikit-learn
!pip install sentence-transformers lightgbm pandas numpy scikit-learn

import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.neighbors import NearestNeighbors

import lightgbm as lgb
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("goa_complete_trainable_dataset.csv")

# Clean columns
df = df.loc[:, ~df.columns.duplicated()]

df = df.rename(columns={
    "Place Name": "place_name",
    "Category": "category",
    "Rating": "rating",
    "Model_Training_Text_Feature": "text"
})

# Convert 'rating' to numeric, coercing errors to NaN before filling
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

df['text'] = df['text'].fillna("")
df['category'] = df['category'].fillna("unknown")
df['rating'] = df['rating'].fillna(df['rating'].mean())
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings
text_embeddings = model.encode(df['text'].tolist(), show_progress_bar=True)

text_embeddings = np.array(text_embeddings)
```

```python
from sklearn.compose import ColumnTransformer

categorical_cols = ['category']
numeric_cols = ['rating']

# Preprocessor for categorical and numerical features only
feature_preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', StandardScaler(), numeric_cols)
    ],
    remainder='drop' # Drop other columns not explicitly transformed
)

# Transform categorical and numerical features from the DataFrame
X_features = feature_preprocessor.fit_transform(df)

# Combine with SentenceTransformer text embeddings. Ensure X_features is dense if it's sparse.
X = np.hstack((X_features.toarray() if hasattr(X_features, 'toarray') else X_features, text_embeddings))
print(f"Shape of X: {X.shape}")

# Extract the fitted transformers for later use
ohe = feature_preprocessor.named_transformers_['cat']
scaler = feature_preprocessor.named_transformers_['num']
print(f"Number of categories fitted in OHE: {ohe.categories_[0].shape[0]}")

# Extract place names for later use in recommendations
place_names = df['place_name']
```

```python
knn = NearestNeighbors(n_neighbors=10, metric='cosine')
knn.fit(X)

def get_candidates(user_vector):
    distances, indices = knn.kneighbors(user_vector)
    return indices[0]
```

```python
# Example user history (places visited)
user_history = ["Baga Beach", "Anjuna Beach"]

def get_user_profile(history):
    indices = df[df['place_name'].isin(history)].index
    return np.mean(X[indices], axis=0)
```

```python

train_data = []
group_sizes = []

num_items = len(df)

for i in range(num_items):
    current_group_comparisons = 0
    for j in range(num_items):
        if i != j:
            similarity = (X[i] @ X[j].T).toarray()[0,0]
            label = 1 if df['category'].iloc[i] == df['category'].iloc[j] else 0

            train_data.append([similarity, df['rating'].iloc[j], label])
            current_group_comparisons += 1
    if current_group_comparisons > 0:
        group_sizes.append(current_group_comparisons)

train_df = pd.DataFrame(train_data, columns=['similarity', 'rating', 'label'])

X_train = train_df[['similarity', 'rating']]
y_train = train_df['label']

rank_model = lgb.LGBMRanker()
rank_model.fit(X_train, y_train, group=group_sizes)
```

```python
import joblib

# Save the trained LightGBM model
joblib.dump(rank_model, 'rank_model.pkl')
print("Rank model saved as 'rank_model.pkl'")
```

```python
def recommend_places_advanced(user_input, user_history, top_n=5):

    # Step 1: User embedding
    user_cat = ohe.transform([[user_input['category']]]).toarray()
    user_rating = scaler.transform([[user_input['rating']]])
    user_text_emb = model.encode([user_input['text']])

    # Stack in the same order as X was constructed: categorical, numerical, then text embeddings
    user_vector = np.hstack([user_cat, user_rating, user_text_emb])
    print(f"Shape of user_cat: {user_cat.shape}")
    print(f"Shape of user_rating: {user_rating.shape}")
    print(f"Shape of user_text_emb: {user_text_emb.shape}")
    print(f"Shape of user_vector (before adding history): {user_vector.shape}")

    # Step 2: Add user history profile
    user_profile = get_user_profile(user_history)
    print(f"Shape of user_profile (from get_user_profile): {user_profile.shape}")
    # Reshape user_profile to (1, -1) to match user_vector's shape for element-wise addition
    final_user_vector = (user_vector + user_profile.reshape(1, -1)) / 2
    print(f"Shape of final_user_vector: {final_user_vector.shape}")

    # Step 3: Candidate generation
    candidates = get_candidates(final_user_vector)

    # Step 4: Ranking
    scores = []
    for idx in candidates:
        # Ensure similarity calculation is correct with the new X
        similarity = np.dot(final_user_vector, X[idx])
        # Extract scalar value from similarity array
        similarity_scalar = similarity.item() # Or similarity[0]
        rating = df['rating'].iloc[idx]

        score = rank_model.predict([[similarity_scalar, rating]])[0]
        scores.append((idx, score))

    # Step 5: Sort
    ranked = sorted(scores, key=lambda x: x[1], reverse=True)

    return [df['place_name'].iloc[i[0]] for i in ranked[:top_n]]
```

```python
user_input = {
    "category": "Beach",
    "rating": 4.5,
    "text": "peaceful sunset beach photography"
}

user_history = ["Baga Beach", "Calangute Beach"]

recommend_places_advanced(user_input, user_history)
```
